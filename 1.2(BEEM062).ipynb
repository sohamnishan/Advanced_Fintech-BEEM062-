```python
xt=table['Adj_Close'].values.tolist()
yt=table['close price'].values.tolist()
```

```python
xt = (xt - np.mean(xt))/(np.std(xt))
yt = (yt - np.mean(yt))/(np.std(yt))
```

```python
plt.plot(xt,yt,'go')
plt.xlabel("McDonalds Adj_close_price")
plt.ylabel("Bitcoin")
```

```python
beta = ((np.multiply(yt, xt)).mean()-np.mean(xt)*np.mean(yt))/((np.multiply(xt, xt)).mean()-np.mean(xt)*np.mean(xt))
alpha =np.mean(yt)-beta*np.mean(xt)
print("alpha =",alpha, "\nbeta =",beta)
```

```python
yt_hat = alpha + np.multiply(beta,xt)
L = np.sum(np.multiply(yt - yt_hat,yt - yt_hat))
L
```

```python
def  cal_cost(alpha,beta,xt,yt):
        n = len(yt)
        predictions = alpha + np.multiply(beta,xt)
        cost = (1/2*n) * np.sum(np.square(predictions-yt))
        return cost
cal_cost(alpha,beta,xt,yt)
```

```python
iterations = 50
alpha = -1

loss_history = np.zeros((iterations,1))
alpha_history = np.zeros((iterations,1))

for i in range(iterations):
    alpha = alpha +1
    
    loss_history[i] = cal_cost(alpha,beta,xt,yt)
    alpha_history[i] = alpha
    
plt.plot(alpha_history,loss_history,'go')
```

```python
iterations = 1000
alpha = -1
n = len(yt)

loss_history = np.zeros((iterations,1))
alpha_history = np.zeros((iterations,1))

for i in range(iterations):
    
        
        prediction = alpha + np.multiply(beta,xt)
        residuals = yt-prediction
        residuals_sum = sum(residuals)
        gradient = -(1/n)*0.02*residuals_sum
 
        alpha = alpha - gradient
    
        loss_history[i] = cal_cost(alpha,beta,xt,yt)
        alpha_history[i] = alpha
    
plt.plot(alpha_history,loss_history,'go')
```

```python
iterations = 1000
learningrate = 0.02
beta0 = -1
beta1 = 0
n = len(yt)
index = np.arange(1, iterations+1, 1).reshape((iterations, 1))

loss_history = np.zeros((iterations,1))
beta0_history = np.zeros((iterations,1))
beta1_history = np.zeros((iterations,1))

for i in range(iterations):
       
        prediction = beta0 + np.multiply(beta1,xt)
        residuals = yt-prediction
        residuals_sum = sum(residuals)
        beta0_gradient = -(1/n)*residuals_sum*(learningrate)
        beta0 = beta0 - beta0_gradient
        beta0_history[i] = beta0
    
        residuals_x = np.multiply(residuals,xt)
        residuals_x_sum = sum(residuals_x)
        beta1_gradient = -(1/n)*residuals_x_sum*(learningrate)
        beta1 = beta1 - beta1_gradient
        beta1_history[i] = beta1
    
        loss_history[i] = cal_cost(beta0,beta1,xt,yt)
```

```python
plt.plot(beta1_history,loss_history,'go')
```

```python
plt.subplot(311)
plt.plot(index, beta0_history,'go')
plt.subplot(312)
plt.plot(index, beta1_history,'go')
plt.subplot(313)
plt.plot(index, loss_history,'go')
```
